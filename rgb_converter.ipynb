```python
## importing packages
import os
import numpy as np
from PIL import Image
import cv2
import pathlib
```

```python
## dataset customization
path = '/home/stark/thesis/sram/dataxmc/tflite/dataset/corrupted-dataset/corrupted'

### using Pillow
#dest_path = '/content/drive/MyDrive/tflite/boards/boards3/boards3-mix-RGB-pillow'

## using openCV
dest_path = '/home/stark/thesis/sram/dataxmc/tflite/dataset/corrupted-dataset/corrupted-rgb'
```

```python
# creating folders at destination
folders = []
for foldername in os.listdir(path):
    if foldername.startswith('board'):
        folders.append(foldername)

for folder in folders:
    os.mkdir(os.path.join(dest_path,folder))
    print(folder)
 

    ## file processing
for folder in folders:
    for file in os.listdir(os.path.join(path,folder)):
        if file.startswith('board'):
            ## Pillow
            #image = Image.open(os.path.join(path,folder,file)).convert("RGB")
      
            ## opencv
            image = cv2.imread(os.path.join(path,folder,file), cv2.IMREAD_GRAYSCALE)
            image_rgb = cv2.cvtColor(image, cv2.COLOR_GRAY2RGB)

              # re-write file after conversion
            new_filename = os.path.splitext(file)[0] + '_rgb.png'
            print(new_filename)
      
            ## Pillow
            #image.save(os.path.join(dest_path,folder,new_filename))

              ## Opencv
            cv2.imwrite(os.path.join(dest_path,folder,new_filename), image_rgb)
```
